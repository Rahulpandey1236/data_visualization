```python
#2.. Generate a dataset of random numbers. Visualize the distribution of a numerical variable

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate Random Data
# Generate a dataset of 1000 random numbers from a normal distribution
data = np.random.randn(1000)



fig, axes = plt.subplots(1, 2, figsize=(14, 6))


sns.histplot(data, bins=30, kde=False, ax=axes[0], color='blue')
axes[0].set_title('Histogram of Random Data')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')


sns.kdeplot(data, ax=axes[1], color='red')
axes[1].set_title('Density Plot of Random Data')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Density')


plt.tight_layout()
plt.show()
```

```python
#3.create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create the Dataset
# Create a list of categories
categories = ['A', 'B', 'C', 'D']


np.random.seed(0)
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Value': np.random.randn(100) * 10 + 50
})


plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Value', data=data, estimator=np.mean, ci='sd', palette='viridis')
plt.title('Bar Plot of Categories and Their Corresponding Values')
plt.xlabel('Category')
plt.ylabel('Mean Value with Standard Deviation')
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Value', data=data, palette='viridis')
plt.title('Box Plot of Categories and Their Corresponding Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical 
#variable across different categories.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


categories = ['A', 'B', 'C', 'D']

np.random.seed(0)
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Value': np.random.randn(100) * 10 + 50
})



plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Value', data=data, palette='viridis')
plt.title('Box Plot of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Category', y='Value', data=data, palette='viridis')
plt.title('Violin Plot of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade  categories.

                  
                  
                  
                  
                  
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(15)

# Generating student grades data
grades_data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
grades_df = pd.DataFrame(grades_data)

# Creating a violin plot for student grades
plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=grades_df, palette='muted')
plt.title('Distribution of Scores Across Different Grade Categories')
plt.show()
```

```python
#3.3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(20)


data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)


sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')


plt.figure(figsize=(12, 8))
sns.heatmap(sales_pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Sales Variation Across Different Months and Days')
plt.show()
```

```python
#5.. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


np.random.seed(25)


data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
bubble_sizes = df['Population'] * 10  
plt.scatter(df['GDP'], df['Population'], s=bubble_sizes, alpha=0.5)


for i in range(df.shape[0]):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=12, ha='right')

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs. GDP by Country')
plt.grid(True)
plt.show()

```
